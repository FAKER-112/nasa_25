### Subtask:
Apply PCA to reduce the dimensionality of the one-hot encoded features to two components per original categorical feature.


**Reasoning**:
Create a list of columns that were one-hot encoded, apply PCA to each set of one-hot encoded columns to reduce dimensionality to two components, and concatenate the resulting PCA components with the DataFrame while dropping the original encoded columns.



```python
from sklearn.decomposition import PCA

# 1. Create a list of columns that were one-hot encoded
# This list is derived from the categorical_cols_for_encoding from the previous step
# We need to identify the columns in df_encoded that correspond to these original categorical columns
one_hot_encoded_cols = [col for col in df_encoded.columns if '_' in col and col.split('_')[0] in categorical_cols_for_encoding]

# Create a copy of df_encoded to perform transformations
df_pca = df_encoded.copy()

# Initialize a list to store the names of the original one-hot encoded columns to drop
cols_to_drop = []

# 3. For each original categorical column that was encoded, apply PCA
for original_col in categorical_cols_for_encoding:
    # 2. Identify the corresponding one-hot encoded columns
    current_encoded_cols = [col for col in one_hot_encoded_cols if col.startswith(f'{original_col}_')]

    if current_encoded_cols:
        # 3a. Extract these columns into a separate DataFrame
        df_current_encoded = df_pca[current_encoded_cols]

        # 3b. Instantiate a PCA object
        # Ensure n_components is not greater than the number of features
        n_components = min(2, df_current_encoded.shape[1])
        pca = PCA(n_components=n_components)

        # 3c. Fit the PCA model and transform
        pca_components = pca.fit_transform(df_current_encoded)

        # 3d. Create a new DataFrame with two columns representing the PCA components
        pca_col_names = [f'{original_col}_pca{i+1}' for i in range(n_components)]
        df_pca_components = pd.DataFrame(pca_components, columns=pca_col_names, index=df_pca.index)

        # 3e. Concatenate this new PCA components DataFrame
        df_pca = pd.concat([df_pca, df_pca_components], axis=1)

        # Add the original one-hot encoded columns to the list to drop
        cols_to_drop.extend(current_encoded_cols)

# 3f. Drop the original one-hot encoded columns
df_encoded = df_pca.drop(columns=cols_to_drop)


# 4. Display the first few rows of the updated df_encoded DataFrame and its shape
display(df_encoded.head())
print(f"Shape of the updated DataFrame after PCA: {df_encoded.shape}")
```

**Reasoning**:
The PCA transformation seems to have been applied to the entire DataFrame instead of individual categorical columns. Need to correct the PCA application to process each set of one-hot encoded columns separately and concatenate the results correctly.



```python
from sklearn.decomposition import PCA

# Create a list of columns that were one-hot encoded in the previous step
# This list is derived from the categorical_cols_for_encoding from the previous step
# We need to identify the columns in df_encoded that correspond to these original categorical columns
one_hot_encoded_cols = [col for col in df_encoded.columns if '_' in col and col.split('_')[0] in categorical_cols_for_encoding]

# Create a copy of the original df (before one-hot encoding) to build upon
# This is to avoid carrying over the incorrect PCA application
df_pca = df.copy()

# Initialize a list to store the names of the original one-hot encoded columns to drop later
cols_to_drop_from_original_encoded = []

# For each original categorical column that was encoded, apply PCA
for original_col in categorical_cols_for_encoding:
    # Identify the corresponding one-hot encoded columns in the *previously* encoded dataframe structure
    # We need to work with the structure of the dataframe after the initial one-hot encoding
    current_encoded_cols = [col for col in df_encoded.columns if col.startswith(f'{original_col}_')]

    if current_encoded_cols:
        # Extract these columns from the *initially encoded* dataframe
        df_current_encoded = df_encoded[current_encoded_cols]

        # Instantiate a PCA object
        # Ensure n_components is not greater than the number of features
        n_components = min(2, df_current_encoded.shape[1])
        if n_components > 0:
            pca = PCA(n_components=n_components)

            # Fit the PCA model and transform
            pca_components = pca.fit_transform(df_current_encoded)

            # Create a new DataFrame with columns representing the PCA components
            pca_col_names = [f'{original_col}_pca{i+1}' for i in range(n_components)]
            df_pca_components = pd.DataFrame(pca_components, columns=pca_col_names, index=df_pca.index)

            # Concatenate this new PCA components DataFrame with the df_pca DataFrame
            df_pca = pd.concat([df_pca, df_pca_components], axis=1)

            # Add the original one-hot encoded columns to the list to drop from the *initially encoded* dataframe
            cols_to_drop_from_original_encoded.extend(current_encoded_cols)

# The df_pca now contains the original columns plus the new PCA features.
# We still have the original categorical columns in df_pca at this point.
# We need to drop the original categorical columns that were encoded.
df_pca = df_pca.drop(columns=categorical_cols_for_encoding)


# Display the first few rows of the updated df_pca DataFrame and its shape to verify the transformation.
display(df_pca.head())
print(f"Shape of the updated DataFrame after PCA: {df_pca.shape}")
```

```python
from sklearn.preprocessing import PolynomialFeatures
import numpy as np

# Store original column names before adding new features
original_cols = X_train.columns.tolist()

# 1. Create a new feature 'period_duration_ratio'
X_train['period_duration_ratio'] = X_train['koi_period'] / X_train['koi_duration']
X_val['period_duration_ratio'] = X_val['koi_period'] / X_val['koi_duration']
X_test['period_duration_ratio'] = X_test['koi_period'] / X_test['koi_duration']

# 2. Create a new feature 'depth_period_product'
X_train['depth_period_product'] = X_train['koi_depth'] * X_train['koi_period']
X_val['depth_period_product'] = X_val['koi_depth'] * X_val['koi_period']
X_test['depth_period_product'] = X_test['koi_depth'] * X_test['koi_period']

# 3. Create interaction terms between 'koi_period' and 'koi_duration'
# This is already covered by the product in step 2, but let's make it explicit
# We'll create a new feature name to avoid confusion if the product was intended differently
X_train['period_x_duration'] = X_train['koi_period'] * X_train['koi_duration']
X_val['period_x_duration'] = X_val['koi_period'] * X_val['koi_duration']
X_test['period_x_duration'] = X_test['koi_period'] * X_test['koi_duration']


# 4. Create interaction terms between 'koi_depth' and 'koi_period'
# This is already covered by the product in step 2, no need to repeat

# 5. Create polynomial features for 'koi_period', 'koi_duration', and 'koi_depth' up to degree 2.
# We'll use a separate transformer for clarity and to avoid affecting other features
poly_features = PolynomialFeatures(degree=2, include_bias=False)

# Fit on training data and transform all sets
poly_features.fit(X_train[['koi_period', 'koi_duration', 'koi_depth']])

poly_train = poly_features.transform(X_train[['koi_period', 'koi_duration', 'koi_depth']])
poly_val = poly_features.transform(X_val[['koi_period', 'koi_duration', 'koi_depth']])
poly_test = poly_features.transform(X_test[['koi_period', 'koi_duration', 'koi_depth']])

# Get the names of the new polynomial features
poly_feature_names = poly_features.get_feature_names_out(['koi_period', 'koi_duration', 'koi_depth'])

# Convert polynomial features arrays to DataFrames and concatenate
X_train_poly = pd.DataFrame(poly_train, columns=poly_feature_names, index=X_train.index)
X_val_poly = pd.DataFrame(poly_val, columns=poly_feature_names, index=X_val.index)
X_test_poly = pd.DataFrame(poly_test, columns=poly_feature_names, index=X_test.index)

# Concatenate polynomial features with the main dataframes, dropping original columns used for poly
# We drop the original columns first to avoid duplication before concatenation
X_train = X_train.drop(columns=['koi_period', 'koi_duration', 'koi_depth'])
X_val = X_val.drop(columns=['koi_period', 'koi_duration', 'koi_depth'])
X_test = X_test.drop(columns=['koi_period', 'koi_duration', 'koi_depth'])

X_train = pd.concat([X_train, X_train_poly], axis=1)
X_val = pd.concat([X_val, X_val_poly], axis=1)
X_test = pd.concat([X_test, X_test_poly], axis=1)


# 6. Replace any resulting infinite values with NaN and then impute these NaNs using the median of the respective column in the training set.
X_train.replace([np.inf, -np.inf], np.nan, inplace=True)
X_val.replace([np.inf, -np.inf], np.nan, inplace=True)
X_test.replace([np.inf, -np.inf], np.nan, inplace=True)

# Calculate median from the training set
train_medians = X_train.median()

# Impute NaNs using the calculated medians
X_train.fillna(train_medians, inplace=True)
X_val.fillna(train_medians, inplace=True)
X_test.fillna(train_medians, inplace=True)

# 8. Print the shapes of the updated training, validation, and test feature sets.
print("Shape of updated X_train:", X_train.shape)
print("Shape of updated X_val:", X_val.shape)
print("Shape of updated X_test:", X_test.shape)
```
